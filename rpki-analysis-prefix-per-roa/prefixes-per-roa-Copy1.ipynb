# Prefixes per ROA for a Given Date

## Introduction

A single ROA can contain multiple prefixes and this analysis tries to estimate how much usage of this feature is currently being made by the RPKI community. 

The dataset analyzed is composed of the daily validation output produced by RIPE-NCC's CLI validation software. Available data goes back to January 2013. This notebook averages the amount of prefixes found in ROAs for all RIR trust anchors for a given date.

The 'roadate' variable below can be used to configure the date to be analyzed. Date format is YYYYMMDD or the string 'latest'.

```python
%%capture output
%%bash
roadate=latest
rm DATE\-*
touch DATE-$roadate.md
for rir in lacnic arin apnic-iana afrinic ripe-ncc
do
    fname=$rir.tal-roa-prefixes.csv
    echo "Getting validated ROA data for RIR $rir"
    rm $fname
    wget -N http://ripeval.labs.lacnic.net/rpki/batch-validation/$roadate/$fname
    sort -u -t, -k1 $fname | cut -d, -f1 | uniq -c | sort -rn > $rir-pfx-roa.txt
done
```

## Analisys Script and Results

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#
files = ['apnic-iana-pfx-roa.txt', 'lacnic-pfx-roa.txt', 'arin-pfx-roa.txt', 'ripe-ncc-pfx-roa.txt', 'afrinic-pfx-roa.txt']

## begin
def get_pfx_roa_clean(f):
    f = open(f, "rb")
    rd = csv.reader(f, delimiter=" ")
    i = 0
    pfx_roa = []
    for line in rd:
        # print line[4], line[5][0:20]
        try:
            line = [x for x in line if x != '']
            pfx_roa.append(int(line[0]))
        except:
            print "linea problematica [%s]" % (line[4])
            break
        # pfx_roa.append(5)
        i = i + 1
        if i>100000:
            break
        #
    f.close()
    # print "*** lines %s" % (i)
    return np.array(pfx_roa)
## end get_pfx_roa+stats

for rir in files:
    pr = get_pfx_roa_clean(rir)
    print "Processing RIR data: %s" % (rir)
    print "\tlines: %s" %(len(pr))
    print "\tavg  : %s" % (np.average(pr))
    print " "
    plt.hist(pr, bins=30)
    plt.title("Prefix-ROA distribution for %s" % (rir))
    plt.xlabel("Prefixes per ROA")
    plt.ylabel("ROA Count")
    plt.show()
```
